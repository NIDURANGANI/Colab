# Introduction

Seaborn is a powerful Python data visualization library based on Matplotlib. It provides a high
level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
import pandas as pd
```

# Reading a data set (Vehicle)

```python
data = pd.read_csv('/content/vehicle.csv')
#Using DataFrame in Pandas
df=pd.DataFrame(data)
```

## Cleaning Data using pandas

```python
df.drop_duplicates()
df.head(3)
```

The Price, Mileage, and Capacity columns are converted into numeric data types.

```python
df['Price'] = pd.to_numeric(df['Price'].replace({'Rs ': '', 'Rs.\n': '', ',': ''}, regex=True), errors='coerce')
df['Mileage'] = pd.to_numeric(df['Mileage'].replace({' km': '',
',': ''}, regex=True), errors='coerce')
```

```python
df['Capacity'] = pd.to_numeric(df['Capacity'].replace({' cc':
'', ',': ''}, regex=True), errors='coerce')
```

## Show the selected columns

```python
df[['Brand','Model','Year','Price','Condition','Transmission','Body','Fuel', 'Mileage','Capacity','Review']].head()
```

## Using Seaborn to plot charts

```python
sns.set_style('darkgrid')
sns.histplot(data=df, x='Mileage')
```

```python
sns.set_style('whitegrid')
sns.countplot(data=df, x='Fuel')
```

```python
sns.set_style('darkgrid')
sns.boxplot(x="Review",y="Mileage",data=df)
```

```python
sns.set_style('whitegrid')
sns.scatterplot(data=df, x='Year', y='Price')
```

Heatmap: Correlation Heatmap (Price, Mileage, Year, Capacity)

```python
corr = df[['Price', 'Mileage', 'Year','Capacity']].corr()
sns.heatmap(corr, annot=True)
```

Line Plot: Year vs Price

```python
sns.lineplot(x="Year", y="Price", data=df)
```

Bar Plot: Fuel vs Price with Transmission

```python
sns.barplot(x = 'Fuel', y = 'Price', hue='Transmission', data = df)
```

Swam Plot: Review and Price

```python
sns.set_style('white')
sns.swarmplot(x ="Review", y = "Price", data = df, size = 5)
```
