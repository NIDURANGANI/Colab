# Import pyplot is a module in Matplotlib, Numpy, and Pandas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Read a CSV file

```python
import pandas as pd
data=pd.read_csv("/content/THE World University Rankings 2016-2024.csv", encoding='latin1')
df=pd.DataFrame(data)
df.head()
```

```python
df['International Students'] = pd.to_numeric(df['International Students'].replace({'%': ''}, regex=True), errors='coerce')
df['International Students']
```

# Creating a bar chart

```python
x=df['Name'].head(10)
y=df['Student Population'].head(10)
plt.figure(figsize=(9, 6))
plt.bar(x,y, color='m')
plt.xlabel('Name')
plt.ylabel('Student Population')
plt.title('Name vs Student Population')
plt.xticks(rotation=90)
plt.show()
```

# Creating a line chart

```python
x=df['Name'].head(10)
y1=df['Research Quality'].head(10)
y2=df['Overall Score'].head(10)
plt.figure(figsize=(12, 6))
plt.plot(x,y1, color='b',marker='+')
plt.plot(x,y2, color='r', marker='o')
plt.xlabel('Name')
plt.ylabel('Research Quality and Overall Score')
plt.title('Name vs Research Quality and Overall Score (Top 10 Universities)')
plt.xticks(rotation=90)
plt.legend(['Research Quality','Overall Score'])
plt.grid()
plt.tight_layout()
plt.show()
```

# Creating a histogram

```python
plt.hist(df['Student Population'].head(100), bins=30,
color='blue', edgecolor='black')
plt.title('Histogram')
plt.xticks(rotation=90)
plt.xlabel('Student Population')
plt.ylabel('Frequency')
plt.show()
```

# Creating a pie chart

```python
labels=df['Name'].head(5)
sizes=df['Student Population'].head(5)
plt.figure(figsize=(9, 6))
plt.pie(sizes, labels=labels, autopct='%1.2f%%')
plt.title("Top 5 Universities' Student Population")
plt.show()
```

# Creating a boxplot

```python
plt.boxplot(df['Students to Staff Ratio'].head(10))
plt.title('Boxplot')
plt.ylabel('Students to Staff Ratio')
plt.show()
```

# Creating a scatter plot


```python
x=df['Country'].head(100)
y=df['Research Quality'].head(100)
plt.figure(figsize=(9, 6))
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('Country')
plt.ylabel('Research Quality')
plt.xticks(rotation=90)
plt.show()
```
